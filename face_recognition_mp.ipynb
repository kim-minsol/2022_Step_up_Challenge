```python
import os
import cv2
import numpy as np

model = cv2.face.LBPHFaceRecognizer_create()

faces=[]
ids=[]

dirs = os.listdir("C:/Users/kkhhj/Desktop/face_recognize_data_mp")
for dir_name in dirs:
    dir_path = "C:/Users/kkhhj/Desktop/face_recognize_data_mp/" + dir_name
    id_=int(dir_name[1])
    
    face_dir = os.listdir(dir_path)
    for i in face_dir:
        path = dir_path + "/" + i
        image = cv2.imread(path)
        image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faces.append(image_gray)
        ids.append(id_)

from sklearn.utils import shuffle
X, y = shuffle(faces, ids, random_state=0)

model.train(X, np.array(y))
model.write("trained_model_mp.yml")
print("Finish training model !")
```
